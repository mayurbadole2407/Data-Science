# Objective

###### our main objective is to find those various features which will affect the price of the automobiles

#### Univariate analysis

```python
# scatter plot for each feature
plt.figure(figsize=[19,40])
for i,n in enumerate(num):
    plt.subplot(8,8,i+1)
    plt.scatter(df2[n],np.zeros_like(df2[n]))
    plt.xlabel(n)
    
plt.show()    
```

you can see that the having the scaling of each features.

```python
#cdf of price
count,bin_ed = np.histogram(df2.price,bins=10)
print(count)
print(bin_ed)
pdf = count/(sum(count))
cdf = np.cumsum(pdf)
plt.plot(bin_ed[1:],pdf)
plt.plot(bin_ed[1:],cdf)
```

our price features follows the perito/power law distribution, means that the 80% points are present in front portion of the distribution. It has long tail and not touching to the ground means the prices are descrite random variables.

```python
sb.set_style(style='darkgrid')
plt.figure(figsize=[16,30])
for i,n in enumerate(num):
    plt.subplot(8,8,i+1)
    plt.hist(df2[n],color='red',bins=15)
    plt.xlabel(n)

    
plt.show()    
```

```python
sb.set_style(style='darkgrid')
plt.figure(figsize=[16,30])
for i,n in enumerate(num):
    plt.subplot(8,8,i+1)
    sb.distplot(df2[n],bins=15)
    plt.xlabel(n)

    
plt.show()    
```

1. most of the features are normally distributed as seen 
2. price is slightly log distributed.
